```python
model = KAN(width=[4   ,  3], grid=6, k=3, seed=2024, device=device)  

#def train_acc():
    #return torch.mean((torch.argmax(model(dataset['train_input']), dim=1) == dataset['train_label']).type(dtype))   

def train_acc():
    return torch.mean((torch.argmax(model(dataset['train_input']), dim=1) == torch.argmax(dataset['train_label'], dim=1)).float())


#def test_acc():                           
    #return torch.mean((torch.argmax(model(dataset['test_input']), dim=1) == dataset['test_label']).type(dtype))

def test_acc():
    return torch.mean((torch.argmax(model(dataset['test_input']), dim=1) == torch.argmax(dataset['test_label'], dim=1)).float())


results = model.fit(dataset, opt="LBFGS", steps=20, metrics=(train_acc, test_acc), loss_fn = torch.nn.BCEWithLogitsLoss());
```

```python
# How accurate is this formula ? 
def acc(formula1 , formula2, formula3 , X, y):
    batch = X.shape[0]
    correct = 0
    for i in range(batch):
        # Substitute the values and evaluate numerically
        logit1 = float(formula1.subs({'x_1': X[i, 0], 'x_2': X[i, 1]}).evalf())
        logit2 = float(formula2.subs({'x_1': X[i, 0], 'x_2': X[i, 1]}).evalf())
        logit3 = float(formula3.subs({'x_1': X[i, 0], 'x_2': X[i, 1]}).evalf()) 
        
        # Compare predictions
        correct += (logit2 > logit1) == y[i]
    return correct / batch

# Test the function
print('train acc of the formula:', acc(formula1, formula2, formula3 , dataset['train_input'], dataset['train_label']))
print('test acc of the formula:', acc(formula1, formula2, formula3 , dataset['test_input'], dataset['test_label']))
```

```python
def acc(formula1, formula2, formula3 , X, y):
    batch = X.shape[0]
    correct = 0
    for i in range(batch):
        # Create a substitution dictionary dynamically
        subs_dict = {f'x_{j+1}': X[i, j] for j in range(X.shape[1])}
        
        # Substitute the values into the formulas
        logit1_sym = formula1.subs(subs_dict)
        logit2_sym = formula2.subs(subs_dict)
        logit3_sym = formula3.subs(subs_dict) 
        
        # Debugging: Print intermediate results
        #print(f"Substitution Dictionary: {subs_dict}")
        #print(f"Substituted logit1: {logit1_sym}")
        #print(f"Substituted logit2: {logit2_sym}")
        #print(f"Substituted logit3: {logit3_sym}")  
        
        # Evaluate the symbolic expressions numerically
        logit1 = float(logit1_sym.evalf())
        logit2 = float(logit2_sym.evalf())
        logit3 = float(logit3_sym.evalf()) 
        
        # Determine the predicted class based on the maximum logit
        predicted_class = max((logit1, 0), (logit2, 1), (logit3, 2), key=lambda x: x[0])[1]
        
        # Compare predictions with actual labels
        correct += (predicted_class == y[i]) 
        

    return correct / batch

# Test the function 
print('train acc of the formula:', acc(formula1, formula2, formula3 ,dataset['train_input'], dataset['train_label']))
print('test acc of the formula:', acc(formula1, formula2, formula3 , dataset['test_input'], dataset['test_label']))
```
